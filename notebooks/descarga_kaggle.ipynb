## 0. Requisito previo: credenciales de Kaggle

Para descargar datasets de Kaggle necesitás un token de API:

1. Ir a [kaggle.com](https://www.kaggle.com) → tu perfil → **Settings** → **API** → **Create New Token**.
2. Se descarga un archivo `kaggle.json` con este formato:
   ```json
   {"username": "tu_usuario", "key": "tu_api_key"}
   ```
3. En **Colab**: subir el archivo usando la celda de abajo.
4. En **local**: copiarlo a `~/.config/kaggle/kaggle.json` y darle permisos:
   ```bash
   mkdir -p ~/.config/kaggle
   cp kaggle.json ~/.config/kaggle/
   chmod 600 ~/.config/kaggle/kaggle.json
   ```

````python
# ── Colab: subir kaggle.json ─────────────────────────────────────────────────
from google.colab import files
import os

print('Subí tu archivo kaggle.json:')
uploaded = files.upload()

# Mover al lugar que espera la CLI de Kaggle
os.makedirs(os.path.expanduser('~/.config/kaggle'), exist_ok=True)
os.rename('kaggle.json', os.path.expanduser('~/.config/kaggle/kaggle.json'))
os.chmod(os.path.expanduser('~/.config/kaggle/kaggle.json'), 0o600)

print('✅ Credenciales configuradas.')
# ─────────────────────────────────────────────────────────────────────────────
````

## 1. Instalar Kaggle CLI y descargar el dataset

## 2. Verificar estructura

````python
from pathlib import Path

DEST_DIR = Path(REPO_PATH) / 'data' / 'kaggle_dataset'

for class_dir in ['Oil', 'No_Oil']:
    folder = DEST_DIR / class_dir
    if folder.exists():
        n = len(list(folder.glob('*.jpg'))) + len(list(folder.glob('*.png')))
        print(f'  {class_dir}: {n} imágenes')
    else:
        print(f'  ⚠️  Carpeta no encontrada: {folder}')
        print('     Puede que el ZIP tenga una estructura diferente.')
        print('     Listar contenido del directorio:')
        for p in DEST_DIR.iterdir():
            print(f'       {p.name}')
        break

print('\n✅ Dataset listo. Podés correr entrenamiento_mlp.ipynb.')
````
